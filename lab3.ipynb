```python

import re
import datetime
import pandas as pd

log_path = "setupapi.dev2.log"
usb_devices = []
# Read the contents of the setupapi.dev.log file
with open(log_path, "r") as log_file:
    # Store information about each USB device in a dictionary
    for line in log_file:
        match = re.match(r'^>>>  \[Device Install.*#(Disk&Ven_[A-Za-z0-9]+)&(Prod_([\w\s\S]+?))&(Rev_([\w\s\S]+?))#([\w\s\S]+?)#.*\]', line)
        if match:
            usb_devices.append({
                'device_vendor_id': match.group(1),
                'device_product_id': match.group(2),
                'device_instance_id': match.group(4),
                'serial_number': match.group(6),
                'event_time': next(log_file).split("start")[1].split("\n")[0],
            })

df = pd.DataFrame(usb_devices)

display(df)
```

```python
import os
import pandas as pd
import hashlib
import time

dir_path = "D:\\"

# f = magic.Magic()

file_names = []
original_extensions = []
extension_matches = []
md5s = []
sha1s = []
sha256s = []
magic_numbers = []
creation_time = []
modification_time = []
access_time = []


for file in os.listdir(dir_path):
    if os.path.isfile(os.path.join(dir_path, file)):
        print(file)
        file_name, extension = os.path.splitext(file)
        file_names.append(file_name)
        original_extensions.append(extension)
        md5s.append(hashlib.md5(file.encode('utf-8')).hexdigest())
        sha1s.append(hashlib.sha1(file.encode('utf-8')).hexdigest())
        sha256s.append(hashlib.sha256(file.encode('utf-8')).hexdigest())
        creation_time.append(time.ctime(os.path.getctime(os.path.join(dir_path,file))))
        modification_time.append(time.ctime(os.path.getmtime(os.path.join(dir_path,file))))
        access_time.append(time.ctime(os.path.getatime(os.path.join(dir_path,file))))

df = pd.DataFrame({
    'file_name': file_names,
    'extension': original_extensions,
    'md5': md5s,
    'sha1': sha1s,
    'sha256': sha256s,
    })

df2 = pd.DataFrame({
    "creation_time": creation_time,
    "modification_time": modification_time,
    'access_time': access_time,
})

display(df)
display(df2)
```
